### Criando um modelo

### Teste 1

Naive Bayes

Atributos aparentemente decisivos para a sobrevivência do passageiro:

    • sexo (convertido para números: 1 - feminino e 2 - masculino)
    
    • classe (1ª, 2ª, 3ª)
    
    • cônjuges, irmãos (SibSp)
    
    • pais, filhos (Parch)

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import KFold
from sklearn import metrics

dataset = train.loc[:, ('PassengerId', 'Sex', 'Pclass', 'SibSp', 'Parch', 'Survived')]

#Female = 1 | Male = 2
dataset.loc[:, 'Sex'] = dataset.loc[:, 'Sex'].apply(lambda sex: 1 if sex == 'female' else 2)

#índice causa erro no Naive Bayes
#dataset.set_index('PassengerId', inplace = True)

print(f'Dataset (treino): {dataset.shape}')

dataset.head()
```

```python
model = GaussianNB()

data = dataset.loc[:, dataset.columns[:5]]
target = dataset.loc[:, dataset.columns[5]]
        
model.fit(data, target)
    
data = dataset.loc[:, dataset.columns[:5]]
target = dataset.loc[:, dataset.columns[5]]
    
predicted = model.predict(data)

print(metrics.classification_report(target, predicted))
print(metrics. confusion_matrix(target, predicted))

k = 5
i = 1

print(f'{k} folds\n\n')

kfold = KFold(n_splits = k, shuffle = False)

for i_train, i_test in kfold.split(dataset):
    data = dataset.loc[i_train, dataset.columns[:5]]
    target = dataset.loc[i_train, dataset.columns[5]]
        
    model.fit(data, target)
    
    data = dataset.loc[i_test, dataset.columns[:5]]
    target = dataset.loc[i_test, dataset.columns[5]]
    
    predicted = model.predict(data)
    
    print(f'\n\nFold #{i} | train: {len(dataset.iloc[i_train])}, test: {len(dataset.iloc[i_test])}')
    print(metrics.classification_report(target, predicted))
    print(metrics.confusion_matrix(target, predicted))
    
    i += 1
```

```python
dataset = test.loc[:, ('PassengerId', 'Sex', 'Pclass', 'SibSp', 'Parch')]

#Female = 1 | Male = 2
dataset.loc[:, 'Sex'] = dataset.loc[:, 'Sex'].apply(lambda sex: 1 if sex == 'female' else 2)

print(f'Dataset (teste): {dataset.shape}')

dataset.head()
```

```python
predicted = model.predict(dataset)

predicted
```

```python
test['Survived'] = predicted

test.head()
```

### Teste 3

Naive Bayes

Utilizando todos os atributos, exceto ID e Nome:

    • sexo (convertido para números: 1 - feminino e 2 - masculino)
    
    • classe (1ª, 2ª, 3ª)
    
    • cônjuges, irmãos (SibSp)
    
    • pais, filhos (Parch)
    
    • idade (substituída pela média dos sobreviventes da classe, quando não informada)
    
    • ticket
    
    • fare
       
    • cabin (substituída pela moda dos sobreviventes da classe correspondente, quando não informada)

    • embarked (substituído pela moda dos sobreviventes da classe correspondente, quando não informado)

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

train.head()
```

```python
survivors = train[train.Survived == 1]

survivors.head()
```

```python
train.count()
```

```python
#idade: substituir NA pela média de idade da classe correspondente
medias = survivors.groupby('Pclass').median().reset_index()

idade_media = medias['Age']
idade_media.index = medias['Pclass']

#cabine: substituir pela moda da cabine da classe correspondente
cabin_moda = pd.Series([survivors.loc[survivors.Pclass == 1, 'Cabin'].mode()[0], 
                        survivors.loc[survivors.Pclass == 2, 'Cabin'].mode()[0],
                        survivors.loc[survivors.Pclass == 3, 'Cabin'].mode()[0]])
cabin_moda.index = medias['Pclass']

#A moda é o porto 'S' para todas as classes
embarked_moda = survivors.loc[:, 'Embarked'].mode()[0]

#fare: substituir pela média da classe correspondente
fare_media = medias['Fare']
fare_media.index = medias['Pclass']

medias
```

```python
def tratar_dados(dataframe, idade_media, cabin_moda, embarked_moda):
    #sexo: 1 - feminino; 2 - masculino
    dataframe.loc[:, 'Sex'] = dataframe.Sex.apply(lambda s: 1 if s == 'female' else 2)
    
    #idade: substituir NA pela média de idade da classe correspondente    
    for i in range(1, 4):
        dataframe.loc[dataframe.Pclass == i, 'Age'] = dataframe.loc[dataframe.Pclass == i, 'Age'].fillna(idade_media[i])
    
    #cabine: substituir pela moda da cabine da classe correspondente e substituir texto por categorias numéricas (factorize)
    for i in range(1, 4):
        dataframe.loc[dataframe.Pclass == i, 'Cabin'] = dataframe.loc[dataframe.Pclass == i, 'Cabin'].fillna(cabin_moda[i])

    counts, uniques = pd.factorize(dataframe.Cabin)
    dataframe.loc[:, 'Cabin'] = counts
    
    #embarked: moda é o porto 'S' para todas as classes; substituir texto por categorias numéricas (factorize)
    dataframe.loc[:, 'Embarked'] = dataframe.loc[:, 'Embarked'].fillna(embarked_moda)
    
    counts, uniques = pd.factorize(dataframe.Embarked)
    dataframe.loc[:, 'Embarked'] = counts
    
    #fare: substituir NA pela média da classe correspondente    
    for i in range(1, 4):
        dataframe.loc[dataframe.Pclass == i, 'Fare'] = dataframe.loc[dataframe.Pclass == i, 'Fare'].fillna(fare_media[i])
    
    #ticket: substituir texto por categorias numéricas (factorize)
    counts, uniques = pd.factorize(dataframe.Ticket)
    dataframe.loc[:, 'Ticket'] = counts
    
    return dataframe
```

```python
tratar_dados(survivors, idade_media, cabin_moda, embarked_moda)

survivors.count()
```

```python
dataset = train.copy()

tratar_dados(dataset, idade_media, cabin_moda, embarked_moda)

dataset.head()
```

```python
colunas = ['Sex', 'Age', 'SibSp', 'Parch', 'Pclass', 'Ticket', 'Embarked', 'Fare', 'Cabin', 'Survived']

model = GaussianNB()

data = dataset.loc[:500, colunas[:9]]
target = dataset.loc[:500, colunas[9]]
        
model.fit(data, target)

data = dataset.loc[501:, colunas[:9]]
target = dataset.loc[501:, colunas[9]]

predicted = model.predict(data)

print(metrics.classification_report(target, predicted))
print(metrics.confusion_matrix(target, predicted))
```

```python
k = 50
i = 1

print(f'{k} folds\n\n')

kfold = KFold(n_splits = k, shuffle = False)

for i_train, i_test in kfold.split(dataset):
    data = dataset.loc[i_train, colunas[:9]]
    target = dataset.loc[i_train, colunas[9]]
        
    model.fit(data, target)
    
    data = dataset.loc[i_test, colunas[:9]]
    target = dataset.loc[i_test, colunas[9]]
    
    predicted = model.predict(data)
    
    print(f'\n\nFold #{i} | train: {len(dataset.iloc[i_train])}, test: {len(dataset.iloc[i_test])}')
    print(metrics.classification_report(target, predicted))
    print(metrics.confusion_matrix(target, predicted))
    
    i += 1
```

```python
dataset = test.copy()

tratar_dados(dataset, idade_media, cabin_moda, embarked_moda)

dataset.head()
```

```python
data = dataset.loc[:, colunas[:9]]

predicted = model.predict(data)

predicted
```

```python
test['Survived'] = predicted

test.head()
```

```python
result = test.loc[:, ('PassengerId', 'Survived')]
result.to_csv("gender_submission.csv", sep = ",", index = False)
result
```

#### Resultado
